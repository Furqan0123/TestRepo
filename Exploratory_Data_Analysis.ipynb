# **Histogram plots for key numerical features**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Key features
num_features = ['horsepower', 'engine-size', 'curb-weight', 'highway-mpg']

# 2x2 subplot
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs = axs.flatten()

for i in range(4):
    sns.histplot(df[num_features[i]], kde=True, ax=axs[i], bins=30, color='skyblue')
    axs[i].set_title(f'Distribution of {num_features[i]}')
    axs[i].set_xlabel(num_features[i])
    axs[i].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

**Interpretaion**;

**Horsepower**

The distribution of horsepower is right-skewed, indicating that most vehicles have horsepower between 60 and 120, with a few high-performance cars exceeding 200, acting as outliers.

**Engine Size**

Engine size also shows a right-skewed distribution, with the majority of cars having engines in the 100–130 range, while larger engines (above 200) are less common and indicate specialty or luxury vehicles.

**Curb Weight**

The curb weight distribution is slightly right-skewed, with most vehicles weighing between 2000 and 3000 lbs. Heavier cars above 3500 lbs are rare and likely represent high-end or utility vehicles.

**Highway MPG**

This distribution is approximately normal (bell-shaped), with most cars achieving 25–35 miles per gallon on the highway, suggesting consistent fuel efficiency among standard consumer vehicles.

# **Continuous Numerical Variables:**
Continuous numerical variables are variables that may contain any value within some range. A great way to visualize these variables is by using scatterplots with fitted lines.

In order to start understanding the (linear) relationship between an individual variable and the price, we can use "regplot" which plots the scatterplot plus the fitted regression line for the data. This will be useful later on for visualizing the fit of the simple linear regression model as well.

**Let's find the scatterplot of "engine-size" and "price".**

```python
sns.regplot(x="engine-size", y="price", data=df, line_kws={"color": "red"})
plt.ylim(0,)
```

As the engine-size goes up, the price goes up: this indicates a positive direct correlation between these two variables. Engine size seems like a pretty good predictor of price since the regression line is almost a perfect diagonal line.

```python
#We can examine the correlation between 'engine-size' and 'price'.
df[["engine-size", "price"]].corr()
```

We can examine the correlation between 'engine-size' and 'price' and see that it's approximately 0.87.

# **Grouping**

The "groupby" method groups data by different categories. The data is grouped based on one or several variables, and analysis is performed on the individual groups.

For example, let's group by the variable "drive-wheels". We see that there are 3 different categories of drive wheels.

```python
df['drive-wheels'].unique()
```

```python

df_group_one = df[['drive-wheels','body-style','price']]

#We can then calculate the average price for each of the different categories of data.
# grouping results
df_grouped = df_group_one.groupby(['drive-wheels'], as_index=False).agg({'price': 'mean'})
df_grouped
```

It seems rear-wheel drive vehicles are, on average, the most expensive, while 4-wheel and front-wheel are approximately the same in price.

let's group by both 'drive-wheels' and 'body-style'. This groups the dataframe by the unique combination of 'drive-wheels' and 'body-style'. We can store the results in the variable 'grouped_test1'.

```python
# grouping results
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

This grouped data is much easier to visualize when it is made into a pivot table. We can convert the dataframe to a pivot table using the method "pivot" to create a pivot table from the groups.

```python
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

Often, we won't have data for some of the pivot cells. We can fill these missing cells with the value 0, but any other value could potentially be used as well. It should be mentioned that missing data is quite a complex subject and is an entire course on its own

```python
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

**Variables: Drive Wheels and Body Style vs. Price**

Let's use a heat map to visualize the relationship between Body Style vs Price.

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

The heatmap plots the target variable (price) proportional to colour with respect to the variables 'drive-wheel' and 'body-style' on the vertical and horizontal axis, respectively. This allows us to visualize how the price is related to 'drive-wheel' and 'body-style'.

# **Correlation and Causation**

**Pearson Correlation**

The Pearson Correlation measures the linear dependence between two variables X and Y.

The resulting coefficient is a value between -1 and 1 inclusive, where:

1: Perfect positive linear correlation.

0: No linear correlation, the two variables most likely do not affect each other.

-1: Perfect negative linear correlation.

```python
df.select_dtypes(include=['number']).corr()
```

**P-value**

What is this P-value? The P-value is the probability value that the correlation between these two variables is statistically significant. Normally, we choose a significance level of 0.05, which means that we are 95% confident that the correlation between the variables is significant.

By convention, when the

p-value is

 0.001: we say there is strong evidence that the correlation is significant.
the p-value is

 0.05: there is moderate evidence that the correlation is significant.
the p-value is

 0.1: there is weak evidence that the correlation is significant.
the p-value is

 0.1: there is no evidence that the correlation is significant.

```python
#We can obtain this information using "stats" module in the "scipy" library.
from scipy import stats
```

Let's calculate the Pearson Correlation Coefficient and P-value of 'wheel-base' and 'price'.

```python
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```

Since the p-value is 0.001, the correlation between wheel-base and price is statistically significant, although the linear relationship isn't extremely strong (~0.585).

## **Insights and Interpretation**

### 1. **Engine Size, Curb Weight, and Horsepower Strongly Predict Price**

* There is a **strong positive correlation** between `engine-size` and `price` (correlation coefficient ≈ **0.87**), making it one of the best predictors.
* `curb-weight` and `horsepower` also show strong relationships with price (correlations ≈ **0.83** and **0.81** respectively).
* **Interpretation:** Larger engines and heavier vehicles typically belong to more expensive car categories like SUVs or luxury sedans.

---

### 2. **Negative Correlation Between Fuel Efficiency and Price**

* `highway-mpg` has a **negative correlation** (≈ **-0.70**) with `price`.
* **Interpretation:** High fuel efficiency is generally seen in smaller, more economical cars, which are less expensive.

---

### 3. **Visual Patterns From Histogram Plots**

* The **distribution of horsepower**, **engine-size**, and **curb-weight** is **right-skewed**, indicating a concentration of typical consumer cars with a few high-end models (outliers).
* **Highway-mpg** is nearly **normally distributed**, reflecting a balanced representation of fuel efficiency across vehicles.
* **Interpretation:** Most vehicles in the dataset are average consumer vehicles, with only a few extremes like high-performance or ultra-efficient models.

---

### 4. **Drive-Wheel Type Impacts Price**

* Grouped analysis shows **rear-wheel drive (RWD)** vehicles are significantly more expensive on average than **front-wheel drive (FWD)** or **4WD** cars.
* **Interpretation:** This aligns with sports and luxury cars (often RWD) being priced higher than regular sedans and hatchbacks (usually FWD).

---

### 5. **Outliers and Imbalanced Categories**

* Distribution and boxplot visualizations reveal **outliers** in features like `horsepower` and `curb-weight`.
* Some categorical variables (e.g., `engine-location`) are heavily imbalanced — nearly all cars are **front-engine**.
* **Interpretation:** This limits the statistical reliability of such variables and may skew model training if not handled carefully.

---

### 6. **Body Style Alone Is Not a Reliable Predictor**

* Although price varies slightly with `body-style`, the overlap is large in boxplots.
* **Interpretation:** Body style alone does not strongly determine price — it's more meaningful when combined with drive-wheels or engine features.

---

The EDA confirms that features like engine size, horsepower, and weight are strong price indicators, while others like body style and engine location are weak predictors.
